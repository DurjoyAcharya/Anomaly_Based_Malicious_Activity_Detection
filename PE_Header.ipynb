```python
import os
import pandas
import numpy
import pickle
import pefile
import sklearn.ensemble as ek
from sklearn.feature_selection import SelectFromModel
import joblib
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn import svm
import sklearn.metrics as metrics
```

```python
# Feature
X = load_ds.drop(['Name','md5','legitimate'],axis=1).values    #Droping this because classification model will not accept object type elements (float and int only)
# Target variable
y = load_ds['legitimate'].values
```

```python
extratrees = ek.ExtraTreesClassifier().fit(X,y)
model = SelectFromModel(extratrees, prefit=True)
X_new = model.transform(X)
nbfeatures = X_new.shape[1]
```

```python
#Number of important features
nbfeatures
```

```python
#splitting the data (70% - training and 30% - testing)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_new, y ,test_size=0.29, stratify = y)
```

```python
features = []
index = numpy.argsort(extratrees.feature_importances_)[::-1][:nbfeatures]
```

```python
for f in range(nbfeatures):
    print("%d. feature %s (%f)" % (f + 1, load_ds.columns[2+index[f]], extratrees.feature_importances_[index[f]]))
    features.append(load_ds.columns[2+f])
```

```python
model = { "DecisionTree": DecisionTreeClassifier(max_depth=10),
         "RandomForest":ek.RandomForestClassifier(n_estimators=50)}
```

```python
results = {}
for algo in model:
    clf = model[algo]
    clf.fit(X_train,y_train)
    score = clf.score(X_test,y_test)
    print ("%s : %s " %(algo, score))
    results[algo] = score
```

```python
import numpy as np

# Get the indices of the columns in X_test that match the features
feature_indices = np.where(np.isin(load_ds.columns, features))[0]

# Use the indices to predict and calculate the confusion matrix
y_pred = clf.predict(X_test[:, feature_indices])
cm = confusion_matrix(y_test, y_pred)

# Print the confusion matrix
print(f"\nConfusion Matrix for {algo}:\n{cm}")
```

```python
# Algorithm comparison graph (using bar chart)
algorithms = list(results.keys())
scores = list(results.values())
plt.figure(figsize=(8, 6))
plt.bar(algorithms, scores)
plt.xlabel("Algorithms")
plt.ylabel("Accuracy Score")
plt.title("Algorithm Comparison")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
winner = max(results, key=results.get)# Selecting the classifier with good result
print("Using", winner, "for classification, with",len(features), 'features.')
```

```python
joblib.dump(model[winner],'classifier.pkl')
open('features.pkl', 'wb').write(pickle.dumps(features))
```
